```python
plt.figure(figsize=(6,4))
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution of Customers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

Insights: As the age increases number of customer is decreasing.

```python
plt.figure(figsize=(6,4))
sns.histplot(df["Annual_Premium"], bins=40)
plt.title("Annual Premium Distribution")
plt.xlabel("Annual Premium")
plt.ylabel("Count")
plt.show()
```

Insights: Few customers pay very high premium.

```python
plt.figure(figsize=(6,3))
sns.boxplot(x=df["Annual_Premium"])
plt.title("Annual Premium Outliers")
plt.show()
```

Insight: Most customers pay relatively low to moderate premiums, while a small number pay very high premiums.

```python
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()
```

Insights: There are more numbers of male customers than female customers.

```python
sns.countplot(x="Vehicle_Age", data=df)
plt.title("Vehicle Age Distribution")
plt.show()
```

Insights: Most vehicles are 12 years old.

```python
sns.countplot(x="Vehicle_Damage", data=df)
plt.title("Vehicle Damage Distribution")
plt.show()
```

Insights: Large proportion of vehicles have past damage

```python
sns.countplot(x="Response", data=df)
plt.title("Target Variable Distribution")
plt.show()
```

Insights: Majority of customers are not intrested.

```python
sns.countplot(x="Vehicle_Damage", hue="Response", data=df)
plt.title("Vehicle Damage vs Insurance Response")
plt.show()
```

Insights: Customer with damaged vehicles show much higher interest.

```python
sns.countplot(x="Vehicle_Age", hue="Response", data=df)
plt.title("Vehicle Age vs Response")
plt.show()
```

Insights: Customer with older vehicle have higher interest in taking the insurance.

```python
sns.countplot(x="Gender", hue="Vehicle_Damage", data=df)
plt.title("Gender vs Response")
plt.show()
```

Insights: Male customers have more vehicle damage history than female customer.

```python
sns.boxplot(x="Response", y="Age", data=df)
plt.title("Age vs Response")
plt.show()
```

Insight: Older customers are more likely to be interested in vehicle insurance while the younger can prioritize cost-saving over risk protection.

```python
sns.boxplot(x="Response", y="Annual_Premium", data=df)
plt.title("Annual Premium vs Response")
plt.show()
```

Insight: Annual premium alone is not a strong differentiator of customer response.

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    df[["Age", "Annual_Premium", "Vintage", "Response"]].corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```

Insight: Customer response is not driven by any single numerical variable in a linear manner.

```python
sns.boxplot(x="Response", y="Vintage", data=df)
plt.title("Customer Vintage vs Claim Frequency")
plt.show()
```

```python
sns.histplot(df["Vintage"], bins=30, kde=True)
plt.title("Distribution of Customer Vintage (Days)")
plt.xlabel("Vintage (Days)")
plt.ylabel("Count")
plt.show()
```

Insight: The dataset contains a balanced mix of new and long-term customers.

```python
df["Vintage_Group"] = pd.cut(
    df["Vintage"],
    bins=[0, 100, 200, 300, 400],
    labels=["0-100", "101-200", "201-300", "301-400"]
)
```

```python
claim_rate = df.groupby("Vintage_Group")["Response"].mean()

claim_rate.plot(kind="bar", title="Claim Rate by Customer Vintage")
plt.ylabel("Claim Probability")
plt.xlabel("Customer Vintage (Days)")
plt.show()
```

```python
sns.boxplot(x="Response", y="Vintage", data=df)
plt.title("Customer Tenure vs Claim")
plt.show()
```

```python
sns.countplot(x="Previously_Insured", data=df)
plt.title("Distribution of Previously Insured Customers")
plt.show()
```

Insights: People without previously insured vehicle are less intrested in insurance claims.
